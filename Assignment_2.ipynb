```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load pre-trained model and tokenizer
model_name = "gpt2-medium"  # You can use "gpt2-medium", "gpt2-large", or "gpt2-xl" for larger models
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)
```

```python
# Input text for the model
input_text = "In a faraway land"

# Tokenize the input text
input_ids = tokenizer.encode(input_text, return_tensors='pt')
```

```python
# Generate text with the model
output_ids = model.generate(input_ids, max_length=50, num_return_sequences=1, pad_token_id=tokenizer.eos_token_id)
```

```python
# Decode the generated tokens back into text
output_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)

# Print the generated text
print("Generated Text:\n", output_text)
```
